## Равномерное распределение
Параметры: a, b - концы отрезка.

Плотность распределения:
$$
p(x) =
 \begin{cases}
   \frac{1}{b-a} & x \in [a,b]\\
   0 & x \notin [a,b]\\
 \end{cases}
$$

```python
def uniform_density(x, a , b):
    if a <= x <= b:
        return 1/(b-a)
    else:
        return 0


a = -1
b = 3

sample_size = [1000, 10000, 100000]


for n in sample_size:
    Z = np.random.uniform(a, b, n)
    plt.figure()
    plt.hist(Z,density=True, bins=int(n**0.5));

    plot_function(lambda x: uniform_density(x, a, b), a, b)
```

## Плотность распределения

Функция $p(x)$ является плотностью распределения некоторой случайной величины $\xi$, если:
* $p(x) \geq 0$.
* Площадь под графиком $p(x)$ равна 1.

Вероятность того, что случайная величина $\xi$ попала в отрезок $[l,r]$ равна площади под графиком плотности распределения $p(x)$ на отрезке $[l,r]$:

$$
P(l \leq \xi \leq r) = \int_{l}^{r} p(x) dx
$$

А силу того, что любое число попадает а интервал $(-\infty, +\infty)$, то
$$
P(-\infty \leq \xi \leq +\infty) = \int_{-\infty}^{+\infty} p(x) dx = 1
$$

## Численный поиск интеграла

Визуализации:
* https://demonstrations.wolfram.com/RiemannSums/
* https://www.youtube.com/watch?v=NgLd16Dksrs&ab_channel=Visualization101

```python
def integrate(f, p, q, n=100):
    X = np.linspace(p, q, n)
    intergal = 0
    for i in range(len(X)-1):
        intergal += f(X[i])*(X[i+1]-X[i])
    return intergal

def f1(x):
    return 5

def f2(x):
    return x

def f3(x):
    return x**2

f = f3
a = 1
b = 3

plot_function(f, a, b)
for n in [1,5,10,50,100,500,1000]:
    print(f"{n = }", f"Интеграл = {integrate(f, a, b, n)}")
```

# Математическое ожидание

Математическое ожидание непрерывной случайной величины $\xi$, которая имеет плотность распределения $p(x)$, определяется по формуле:

$$
E(\xi) = \int_{-\infty}^{+\infty} x\cdot p(x) dx
$$

Проверим численно, что математическое ожидание равномерно распределенной на отрезке $[a,b]$ случайной величины равняется $E(\xi) = \frac{a+b}{2}$.

```python
a = 1
b = 3
def my_fun(x):
    return x*uniform_density(x, a, b)

n = 100
I = [integrate(my_fun, -10, 10, i) for i in range(n)]
plt.figure(figsize = (10,8))
plt.plot(I, label = 'Численное значение интеграла');
ev = (a+b)/2
plt.axhline(y=ev, color='r', linestyle='--', label = 'Математическое ожидание');
plt.legend();
```
