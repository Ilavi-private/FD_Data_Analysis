```python
# –ó–∞–¥–∞–µ–º –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã —Ç—Ä–µ—Ö —Ç–æ—á–µ–∫
P1 = (1, 2)  # (x1, y1)
P2 = (3, 5)  # (x2, y2)
P3 = (4, 7)  # (x3, y3)

# –°—Ç—Ä–æ–∏–º –∏ —Ä–∏—Å—É–µ–º –ø—Ä—è–º—É—é, –ø—Ä–æ—Ö–æ–¥—è—â—É—é —á–µ—Ä–µ–∑ P1 –∏ P2
# –£—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø—Ä—è–º–æ–π y = kx + b, –≥–¥–µ k - –Ω–∞–∫–ª–æ–Ω –ø—Ä—è–º–æ–π, b - —Å–≤–æ–±–æ–¥–Ω—ã–π —á–ª–µ–Ω
# –ù–∞—Ö–æ–¥–∏–º k –∏ b –∏–∑ —É—Ä–∞–≤–Ω–µ–Ω–∏—è —Å–∏—Å—Ç–µ–º—ã —É—Ä–∞–≤–Ω–µ–Ω–∏–π —Å –¥–≤—É–º—è —Ç–æ—á–∫–∞–º–∏
A = np.array([[P1[0], 1], [P2[0], 1]])
b = np.array([P1[1], P2[1]])
k, b = np.linalg.solve(A, b)

# –°–æ–∑–¥–∞–µ–º —Ç–æ—á–∫–∏ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –ø—Ä—è–º–æ–π
x = np.linspace(0, 5, 100)
y = k * x + b

# –†–∏—Å—É–µ–º –ø—Ä—è–º—É—é, –ø—Ä–æ—Ö–æ–¥—è—â—É—é —á–µ—Ä–µ–∑ P1 –∏ P2
plt.plot(x, y, label='Line through P1 and P2')

# –°—Ç—Ä–æ–∏–º –∏ —Ä–∏—Å—É–µ–º –ø—Ä—è–º—É—é, –ø—Ä–æ—Ö–æ–¥—è—â—É—é —á–µ—Ä–µ–∑ P1 –∏ P3
A = np.array([[P1[0], 1], [P3[0], 1]])
b = np.array([P1[1], P3[1]])
k, b = np.linalg.solve(A, b)

# –°–æ–∑–¥–∞–µ–º —Ç–æ—á–∫–∏ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –ø—Ä—è–º–æ–π
x = np.linspace(0, 5, 100)
y = k * x + b

# –†–∏—Å—É–µ–º –ø—Ä—è–º—É—é, –ø—Ä–æ—Ö–æ–¥—è—â—É—é —á–µ—Ä–µ–∑ P1 –∏ P3
plt.plot(x, y, label='Line through P1 and P3')

# –°—Ç—Ä–æ–∏–º –∏ —Ä–∏—Å—É–µ–º –ø—Ä—è–º—É—é, –ø—Ä–æ—Ö–æ–¥—è—â—É—é —á–µ—Ä–µ–∑ P2 –∏ P3
A = np.array([[P2[0], 1], [P3[0], 1]])
b = np.array([P2[1], P3[1]])
k, b = np.linalg.solve(A, b)

# –°–æ–∑–¥–∞–µ–º —Ç–æ—á–∫–∏ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –ø—Ä—è–º–æ–π
x = np.linspace(0, 5, 100)
y = k * x + b

# –†–∏—Å—É–µ–º –ø—Ä—è–º—É—é, –ø—Ä–æ—Ö–æ–¥—è—â—É—é —á–µ—Ä–µ–∑ P2 –∏ P3
plt.plot(x, y, label='Line through P2 and P3')

# –†–∏—Å—É–µ–º —Ç–æ—á–∫–∏ P1, P2, P3
plt.scatter([P1[0], P2[0], P3[0]], [P1[1], P2[1], P3[1]], color='red', label='Points')
plt.legend()
plt.show()
```

```python
# –ó–∞–¥–∞–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –¥–ª—è —Ç—Ä–µ—Ö –ø—Ä—è–º—ã—Ö
k1, l1 = 2, 1
k2, l2 = -1, 4
k3, l3 = 0.5, 0

# –ù–∞—Ö–æ–¥–∏–º —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –ø—Ä—è–º—ã—Ö (—Ä–µ—à–∞–µ–º —Å–∏—Å—Ç–µ–º—ã —É—Ä–∞–≤–Ω–µ–Ω–∏–π)
A = np.array([[k1, -1], [k2, -1]])
b = np.array([-l1, -l2])
x, y = np.linalg.solve(A, b)

# –ù–∞—Ö–æ–¥–∏–º —Ç–æ—á–∫—É –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –ø—Ä—è–º–æ–π 1 –∏ 3
x_13 = (l3 - l1) / (k1 - k3)
y_13 = k1 * x_13 + l1

# –ù–∞—Ö–æ–¥–∏–º —Ç–æ—á–∫—É –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è –ø—Ä—è–º–æ–π 2 –∏ 3
x_23 = (l3 - l2) / (k2 - k3)
y_23 = k2 * x_23 + l2

# –°–æ–∑–¥–∞–µ–º –º–∞—Å—Å–∏–≤—ã –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –ø—Ä—è–º—ã—Ö
x_values = np.linspace(-10, 10, 100)
y_values1 = k1 * x_values + l1
y_values2 = k2 * x_values + l2
y_values3 = k3 * x_values + l3

# –†–∏—Å—É–µ–º –ø—Ä—è–º—ã–µ
plt.plot(x_values, y_values1, label='y = {}x + {}'.format(k1, l1))
plt.plot(x_values, y_values2, label='y = {}x + {}'.format(k2, l2))
plt.plot(x_values, y_values3, label='y = {}x + {}'.format(k3, l3))

# –†–∏—Å—É–µ–º —Ç–æ—á–∫–∏ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è
plt.scatter(x, y, color='red', label='Intersection point 1&2')
plt.scatter(x_13, y_13, color='green', label='Intersection point 1&3')
plt.scatter(x_23, y_23, color='blue', label='Intersection point 2&3')

plt.legend()
plt.grid()
plt.show()
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—è –º–∞—Ç—Ä–∏—Ü—ã
def check_determinant(n):
	A = np.zeros((n, n))
	b = np.arange(1, n + 1)
	for i in range(n):
		A[i, i] = 1
		if i == n - 1:
			A[i, 0] = 1
		else:
			A[i, i + 1] = 1

	determinant = np.linalg.det(A)
	return determinant

# –ù–∞–π–¥–µ–º –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ n, –¥–ª—è –∫–æ—Ç–æ—Ä–æ–≥–æ —Å–∏—Å—Ç–µ–º–∞ –∏–º–µ–µ—Ç —Ä–µ—à–µ–Ω–∏–µ

max_n = 3
solvable_n = ''

for n in range(3, 31):
	determinant = check_determinant(n)
	if determinant != 0:
		max_n = n
		solvable_n += f' {str(max_n)}'

# –ù–∞—Ö–æ–¥–∏–º —Ä–µ—à–µ–Ω–∏–µ –¥–ª—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ n
A_max = np.zeros((max_n, max_n))
b_max = np.arange(1, max_n + 1)
for i in range(max_n):
	A_max[i, i] = 1
	if i == max_n - 1:
		A_max[i, 0] = 1
	else:
		A_max[i, i + 1] = 1
x_max = np.linalg.solve(A_max, b_max)

# –û—Ç–æ–±—Ä–∞–∂–∞–µ–º —Ç–æ—á–∫–∏ (i, x_i) –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ
plt.scatter(np.arange(1, max_n + 1), x_max, color='b')
plt.xlabel('i')
plt.ylabel('x_i')

print(f'–°–∏—Å—Ç–µ–º–∞ –∏–º–µ–µ—Ç —Ä–µ—à–µ–Ω–∏–µ –ø—Ä–∏  ùëõ  - {solvable_n}')

solution = [f'x_{i} = {x_max[i]}' for i in range(len(x_max))]
print('–ú–Ω–æ–∂–µ—Å—Ç–≤–æ —Ä–µ—à–µ–Ω–∏–π –¥–ª—è —Å–∏—Å—Ç–µ–º—ã —Å n = {}:'.format(max_n))
for s in solution:
	print(s)
	
plt.title(f'–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ n (3 ‚â§ n ‚â§ 30), –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º —Å–∏—Å—Ç–µ–º–∞ –∏–º–µ–µ—Ç —Ä–µ—à–µ–Ω–∏–µ —Ä–∞–≤–Ω–æ {max_n}')
plt.grid()
plt.show()

#–í—ã–≤–æ–¥–∏–º –≤—Å–µ –≤–æ–∑–º–æ–∂–Ω—ã–µ n


# –í—ã–≤–æ–¥–∏–º –º–Ω–æ–∂–µ—Å—Ç–≤–æ —Ä–µ—à–µ–Ω–∏–π –¥–ª—è —Å–∏—Å—Ç–µ–º—ã

	
```
