```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
n = 40
d = 6
X = np.arange(n)
Y = np.array([i**d for i in range(n)])
A = np.ones(d+1) #Коэффициенты полинома степени d равны 1


# Функция для вычисления расстояния R для полинома произвольной степени d
def get_distance(X, Y, A):
    R = 0
    for i in range(len(X)):
        r = np.polyval(A, X[i]) - Y[i] # Вычисляем значение полинома A в точке X[i] и вычитаем Y[i]
        R += r*r
    return math.sqrt(R / len(X))

# Пример использования
print('R =', get_distance(X, Y, A))

linsp  = np.linspace(X.min(), X.max())
f = np.poly1d(np.flip(A)) # Создаем полином при помощи встроенной функции np.poly1d
fun = [f(x) for x in linsp] # Создаем точки для построения графика
plt.plot(linsp,fun)
plt.plot(X,Y, 'ro')
plt.grid(True);
```

```python
def get_poly(X, Y):
    # Формирование матрицы A для метода наименьших квадратов
    A = np.array([X**2, X, np.ones_like(X)]).T

    # Решение системы уравнений методом наименьших квадратов
    P = A.T @ A
    Q = A.T @ Y
    R = np.linalg.solve(P, Q)

    print('Коэффициенты полинома:', R)
    f = np.poly1d(R)  # Создаем полином при помощи встроенной функции np.poly1d
    return f

def plot_picture(X, Y, f):
    plt.plot(X, Y, 'ro')
    linsp = np.linspace(min(X), max(X))
    fun = [f(x) for x in linsp]  # Создаем точки для построения графика
    plt.plot(linsp, fun)
    plt.axis('equal')
    plt.grid(True)

X = np.array([0, 0, 3, 3, 5])
Y = np.array([-1, 1, 4, 2, 2])
f = get_poly(X, Y)
plot_picture(X, Y, f)
```

```python
def get_poly(X, Y, d):
    # Формирование матрицы A для метода наименьших квадратов
    A = np.vander(X, d+1)

    # Решение системы уравнений методом наименьших квадратов
    P = A.T @ A
    Q = A.T @ Y
    R = np.linalg.solve(P, Q)

    print('Коэффициенты полинома:', R)
    f = np.poly1d(R)  # Создаем полином при помощи встроенной функции np.poly1d
    return f

def plot_picture(X, Y, f, d):
    plt.plot(X, Y, 'ro')
    linsp = np.linspace(min(X), max(X))
    fun = [f(x) for x in linsp]  # Создаем точки для построения графика
    plt.plot(linsp, fun, label=f"deg = {d}")
    plt.axis('equal')
    plt.grid(True)
    plt.legend()

n = 20
X = np.arange(n)
Y = np.sin(2*X) + X**0.5

for d in range(1, n):
    f = get_poly(X, Y, d)
    plot_picture(X, Y, f, d)
    plt.show()
```
