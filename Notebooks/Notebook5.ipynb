# Случайные величины и выборки

Визуализация: https://seeing-theory.brown.edu/probability-distributions/index.html

random_variable_simple(A,P) -- Пример случайной величины принимающей 3 значения A[0], A[1] и A[2] с вероятностями P[0], P[1] и P[2]

```python
def random_variable_simple(A,P):
    '''
    A - Список значений
    P - Список вероятностей
    len(A) = len(P) = 3
    '''
    x = rnd.random() # Возвращает случайное значение равновероятно из полуинтервала [0,1)
    if x < P[0]: return A[0]
    elif P[0] <= x < P[0] + P[1]: return A[1]
    else: return A[2]
```

```python
def plot_frequency(X):
    labels, counts = np.unique(X, return_counts=True) # Находим количество значений каждого типа
    plt.bar(labels, counts/len(X), align='center', alpha=0.5) # Рисуем соответствующие столбики
```

```python
A = np.array([-1, 2,3])
P = np.array([0.3, 0.2, 0.5])
n = 100
R = np.array(np.arange(n))
X = np.array([random_variable_simple(A,P) for _ in R])
print(X)
plt.figure(0)
plt.scatter(X,R,marker='.'); # Рисуем выборку в виде точек
plt.figure(1)
plot_frequency(X)
```

# Математическое ожидание случайной величины
### В теории вероятности

Пусть дискретная случайная величина $\xi$ принимает значения $a_1, \dots, a_n$ с вероятнстями $p_1, \dots, p_n$.

Математическим ожиданием $\xi$  называется число:
$$
E(\xi) = p_1a_1 + p_2a_2 + \dots + p_na_n = \sum_{i=1}^n p_i \cdot a_i =
\begin{pmatrix}
p_1 & p_2 & \dots & p_n
\end{pmatrix}
\begin{pmatrix}
a_1 \\
a_2 \\
\vdots \\
a_n
\end{pmatrix}
$$

### В статистике

Пусть $x_1, \dots x_n$  - выборка значений.

Средним значением называется число:
$$
\overline{x} = \frac{1}{n} \sum_{i=1}^nx_i
$$

## Стандартное отклонение

### В терии вероятностей
__Среднеквадратичным отклонением__ случайной величины называют средний разброс ее значений относительно математического ожидания
$$
\sigma(\xi) = \sqrt{E((\xi - E(\xi))^2)}
$$

Если $\xi$ - случайная величина, то $\xi_1 = \xi - E(\xi)$ - тоже случайная величина с математическим ожиданием равным 0.  Если $\xi_1$ - случайная величина, то $\xi_2 = (\xi_1)^2$ - тоже случайная величина. Математическое ожидание $\xi_2$ - это число (его называют дисперсией $\xi$). Корень из этого числа - тоже число, называемое стандартным отклонением случайной величины $\xi$. Таким образом все, что необходимо уметь для подсчета стандартного отклонения - это умение находить математические ожидания различных случайных величин.

### В статистике
Пусть $X = (x_1, \dots, x_n)$ - выборка. Тогда __стандартное отклонение__ X - это

$$
\sigma(X) = \sqrt{\frac{1}{n} \cdot \sum_{i=1}^{n}(x_i - \overline{X})^2},
$$
где $\overline{X} = \frac{x_1 + \dots + x_n}{n}$.

```python
k = 20 # Количество значений
n = 100 # Размер выборки
A = np.arange(1,k+1) # 1,2,3, ..., k
P = np.arange(1,k+1) # 1,2,3, ..., k
P = P / sum(P) # Нормируем, чтобы в сумме получилась 1.
X = np.random.choice(A, size=n, p=P)

expected_value = A.dot(P) # Математическое ожидание (Expected Value)
standard_deviation = (((A - expected_value)**2).dot(P))**0.5 # Считаем среднеквадратичное отклонение по математичекой формуле

mean = X.mean()
std = X.std() # Считаем среднеквадратичное отклонение при помощи встроенной функции

print("Математическое ожидание = %.5f" % expected_value)
print("Среднее значение выборки = %.5f" % mean)
print("Отклонение случайной величины = %.5f" % standard_deviation)
print("Отклонение выборки = %.5f" % std)

plt.figure(figsize = (10,8))

plot_frequency(X)

plt.axvline(expected_value, alpha=0.5, color="g", label="Математические ожидание")
plt.axvline(expected_value + standard_deviation, alpha=0.5, color="g", linestyle='--', label="Отклонение случайной величины")
plt.axvline(expected_value - standard_deviation, alpha=0.5, color="g", linestyle='--')

plt.axvline(mean, alpha=0.5 , color="r", label="Среднее значение")
plt.axvline(mean + std, alpha=0.5, color="r", linestyle='--', label="Отклонение выборки")
plt.axvline(mean - std, alpha=0.5, color="r", linestyle='--')

plt.legend();
```
